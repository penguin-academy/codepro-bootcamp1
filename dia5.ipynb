Crear una clase animales que tenga atributo de clase y de instancia.

    Atributos de clase: especie.

```python
# Creamos la clase 'animales'

class Animales:

    # Definimos atributos de clase

    especie = 'Reptil'

    # Definimos atributos de instancia (bloque constructor)

    def __init__(self, especie, nombre, color):

        self.especie = especie
        self.nombre = nombre 
        self.color = color

def imprimir(nombre, especie, color):

    print(f'Me llamo {nombre} y soy un {especie.lower()} de color {color}.')

# Creamos objetos

reptil1 = Animales('reptil', 'Rango', 'verde')
reptil2 = Animales('reptil', 'Bobby', 'gris')
reptil3 = Animales('reptil', 'Tony', 'azul')

# Imprimimos los atributos de instancia nombre y color

imprimir(reptil1.nombre, Animales.especie, reptil1.color)
imprimir(reptil2.nombre, Animales.especie, reptil2.color)
imprimir(reptil3.nombre, Animales.especie, reptil3.color)
```

# Herencia de clases

```python
# Herencia de clases

# Pasa de padre a hijo

class Padre:

    nacionalidad = 'Paraguaya'
    
    def __init__(self, nombre, apellido, patrimonio):

        self.nombre = nombre
        self.apellido = apellido
        self.patrimonio = patrimonio

    def presentarse(self):

        print(f'Hola soy {self.nombre}, mucho gusto.')

class Hijo(Padre):

    deudas = 5000

    def __init__(self, nombre, apellido, patrimonio):

        self.nombre = nombre

        super().__init__(nombre, apellido, patrimonio) # Para heredar, se eligen los atributos que queremos

padre1 = Padre('Jose', 'Perez', '100.000')
padre2 = Padre('Juan', 'Paez', '10.000.000')

hijo1 = Hijo('Roman', 'Perez', '100.000')

print(hijo1.nombre)
print(hijo1.nacionalidad)
print(hijo1.deudas)

padre1.presentarse()
```

```python
class Reptil:

    tipo = 'Reptil'

    def __init__(self, color, sonido):

        self.color = color
        self.sonido = sonido
        
    def presentarse(self):

        print(f'Buenos dias, soy un reptil de color {self.color} y hago {self.sonido} cuando hablo.')

class Serpiente(Reptil):

    veneno = 10

    def arrastrarse(self):

        print('Estoy arrastrandome')

class Dragon(Reptil):

    garras = 4

    def escupir_fuego(self):

        print('fUeGoOoOoOoOo')

mi_serpiente = Serpiente('amarillo', 'tsssssss')
mi_piton = Serpiente('azul', 'jajaja')
mi_lagarto = Reptil('negro', 'booom')
mi_dragon = Dragon('negro', 'WAAAAAAAAAA')

mi_serpiente.presentarse()
mi_piton.presentarse()
mi_lagarto.presentarse()

mi_piton.arrastrarse()

mi_dragon.escupir_fuego()
```

# Challenge 3ðŸ¤º

Definir una clase madre que herede a sus clases hijas dos atributos de instancia. Mostrar ambos atributos en pantalla desde los objetos de las clases hijas.

ObservaciÃ³n: Editar las clases hijas para que no se inicialicen.

```python
# Creamos clase 'Madre'

class Madre:

    # Definimos atributos de instancia

    def __init__(self, apellido, color_ojos, tez):

        self.apellido = apellido
        self.color_ojos = color_ojos
        self.tez = tez

    # Creamos un metodo 

    def presentarse(self):

        print(f'Mi apellido es {self.apellido}, tengo piel {self.tez} y ojos de color {self.color_ojos}.')

# Creamos clase 'Hija' que herede atributos de 'Madre'

class Hija(Madre):

    # Definimos un atributo de clase

    hambre = True

    # Creamos un metodo

    def hambre(self):

        print('Tengo hambre.')

hija1 = Hija('Johnson', 'azul', 'blanca')
hija2 = Hija('Aragon', 'verde', 'morocha')

hija1.presentarse()
hija2.presentarse()

hija1.hambre()
```

# Challenge 4ðŸ¤º

Crear dos personajes de una historia que se representen a traves de clases. Ejemplo: Un heroe y un villano. Cada personaje debe tener poderes o acciones diferentes. Y deben presentarse en una frase.

Ejemplo: Yo me llamo Pepito y tengo super fuerza. Ejemplo: Yo me llamo Juanito y puedo volar.

```python
# Creamos clase 'Heroe'

class Heroe():

    valentia = True

    # Definimos atributos de instancia

    def __init__(self, nombre, origen, superpoder):

        self.nombre = nombre
        self.origen = origen
        self.superpoder = superpoder

# Creamos clase 'Flash', heredamos atributos de instancia de 'Heroe'

class Flash(Heroe):

    # Creamos un metodo

    def supervelocidad(self):

        print(f'Mi nombre es {self.nombre}. Vengo de {self.origen}. Puedo correr a {self.superpoder} de kilometros por hora')

# Creamos clase 'Villano'

class Villano:

    bondad = False

    # Creamos atributos de instancia

    def __init__(self, nombre, odio, arma):

        self.nombre = nombre
        self.odio = odio
        self.arma = arma

# Creamos clase 'Voldemort', heredamos atributos de instancia de 'Villano'

class Voldemort(Villano):

    # Creamos un metodo

    def avada_kedavra(self):

        print(f'Me llamo {self.nombre}. Odio a los {self.odio}. Disparo avada kedavra con mi {self.arma} para acabar con ellos.')

# Creamos los objetos

flash = Flash('Barry Allen', 'Ciudad Central', '1080 millones')
voldemort = Voldemort('Voldemort', 'muggles', 'varita')

# Llamamos los metodos

flash.supervelocidad()
voldemort.avada_kedavra()
```

# Diccionarios

```python
# Esto haciamos con las listas

mi_contacto = ['Lucas', 'Valdez', '24', '1.79']
estatura = mi_contacto[3]

print(estatura)

# Los diccionarios funcionan similarmente a las listas, pero no tienen orden
# Se definen por la clave que ponemos

# Creando un diccionario / JSON 

mi_contacto1 = {'Nombre': 'Lucas', 
                'Apellido': 'Valdez',
                'Edad': 24,
                'Estatura': 1.79}

# Imprimo todo el diccionario

print(mi_contacto1)

# Imprimiendo un valor llamando a la clave

print(mi_contacto1['Nombre'])

# Imprimir las claves del diccionario

print(mi_contacto1.keys())

# Imprimir los valores del diccionario

print(mi_contacto1.values())

# Imprimir los items del diccionario

print(mi_contacto1.items())

# Sobreescribir el valor de una clave

mi_contacto1['Nombre'] = 'Arturo'

# AÃ±ade una clave y valor en caso de que no exista

mi_contacto1['Peso'] = 87

# Elimina el ultimo item, popitem() no toma argumentos

mi_contacto1.popitem()

# Eliminar valor y clave especifico

del mi_contacto1['Nombre']

print(mi_contacto1)
```

# Challenge 5 ðŸ¤º

Escribir un programa que cree un diccionario vacio y lo vaya llenando con informacion sobre una persona (por ejemplo nombre, edad, sexo,telefono, correo electronico, etc.) que se le pida al usuario. Cada vez que se aÃ±ada un nuevo dato debe imprimirse el contenido del diccionario.

```python
# Creamos una lista con claves pero 'sin datos'

datos = {'Nombre': '',
         'Edad': '',
         'Sexo': '',
         'Telefono': '',
         'Correo': ''}

# Pedimos al usuario que ingrese los datos, imprimimos luego de cada entrada

datos['Nombre'] = input('Ingresa un nombre: ')
print(datos)
datos['Edad'] = input('Ingresa una edad: ')
print(datos)
datos['Sexo'] = input('Ingresa un sexo: ')
print(datos)
datos['Telefono'] = input('Ingresa un telefono: ')
print(datos)
datos['Correo'] = input('Ingresa un correo: ')
print(datos)
```

```python
# Creamos un diccionario 'vacio'

datos = {'Nombre': '',
         'Edad': '',
         'Sexo': '',
         'Telefono': '',
         'Correo': ''}

# Bucle para iterar por todo el diccionario e ingresar los datos del usuario

for d in datos:

    # Reemplazamos el valor de la clave

    datos[d] = input(f'Ingrese su {d}')

    # Imprimimos el diccionario cada que vez que el usario hace una entrada

    print(datos)
```

# Challenge 6 ðŸ¤º 

Crear una tarjeta personal. El usuario (mediante input) puede elegir:

    Crear datos nuevos
    Modificar datos
    Eliminar datos

Ejemplo Tutu

```python
tarjeta_personal = {}

while True:

    menu = int(input('''Selecciona una opcion: 
                    1. Introducir datos
                    2. Modificar datos
                    3. Eliminar datos
                    4. Salir'''))

    if menu == 1:

        clave = input('Ingrese el valor de la clave: ')
        valor = input(f'Ingrese el valor de {clave}')

        tarjeta_personal[clave] = valor

        print(tarjeta_personal)

    elif menu == 2:

        clave = input('Ingrese el valor de la clave a editar: ')
        valor = input(f'Ingrese el nuevo valor de {clave}')

        tarjeta_personal[clave] = valor

        print(tarjeta_personal)

    elif menu == 3:

        clave = input('Ingrese el valor de la clave a eliminar: ')
        del tarjeta_personal[clave]
        
        print(tarjeta_personal)

    elif menu == 4:

        break

    else:

        print('Valor incorrecto.')
```
